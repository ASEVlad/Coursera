```python
import json
import bz2
import numpy as np
import pandas
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import GradientBoostingClassifier
import datetime
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
```

```python
features = pandas.read_csv('D:/Prog/ML/Coursera/Yandex rus/last task/features.csv', index_col='match_id', dtype=np.float32)

features.head()
```

```python
features = features.drop(['duration', 'tower_status_radiant', 'tower_status_dire', 'barracks_status_radiant', 'barracks_status_dire'], axis=1)
features = features.fillna(features.mean())
```

```python
y = features['radiant_win']
X = features.drop('radiant_win', axis=1)
```

```python
scores = list()
k_fold = KFold(n_splits=5, shuffle=True)
for n_estimators in range(10,11,10):
    clf = GradientBoostingClassifier(n_estimators=n_estimators, verbose=True)
    clf.fit(X, y)
    score = cross_val_score(clf, X, y, scoring='roc_auc', cv=k_fold)
    scores.append(score.mean())
scores
```

```python
param_grid  = {'C': np.logspace(-3, -1, 10)}
clf_grid = GridSearchCV(LogisticRegression(random_state=241), param_grid, cv=k_fold, verbose=True, scoring='roc_auc')
clf_grid.fit(X, y)

lr=LogisticRegression(n_jobs=-1,random_state=241,**clf_grid.best_params_)
lr.fit(X, y)
score = cross_val_score(clf, X, y, scoring='roc_auc', cv=k_fold)
print(score.mean())
```
